```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.io import wavfile

# Load the audio files
file1 = 'C:/Users/vesha/PycharmProjects/pC/HOH/noisy/try-01.wav'  # Replace with the filename of the first WAV file
file2 = 'C:/Users/vesha/PycharmProjects/pC/HOH/clean/try-01.wav'  # Replace with the filename of the second WAV file

# Load the audio data and sampling rate for the first file
sr1, signal1 = wavfile.read(file1)

# Load the audio data and sampling rate for the second file
sr2, signal2 = wavfile.read(file2)

# Create a time axis
time1 = np.arange(len(signal1)) / sr1
time2 = np.arange(len(signal2)) / sr2

# Create the figure and axis objects
fig, ax = plt.subplots()

# Plot the first audio signal
ax.plot(time1, signal1, color='blue', label='noisy signal')

# Plot the second audio signal
ax.plot(time2, signal2, color='red', label='clean signal')

# Add labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Amplitude')
ax.set_title('Audio Waveform Comparison')

# Add a legend
ax.legend()

# Display the plot
plt.show()
```
